Don't want the LED on? Turn it off by switching the value from `True` to `False`

```python
cpx_led.value = False
```

```python
import board
import digitalio
import time
import random

print("Hello World!")

class LEDMatrix5x5:
    # GPIO mapping from README
    # An LED is Active is with Column High and Row Low
    COL_PINS = [board.GP2, board.GP3, board.GP4, board.GP5, board.GP25]
    ROW_PINS = [board.GP7, board.GP8, board.GP9, board.GP21, board.GP22]

    def __init__(self):
        self.cols = [digitalio.DigitalInOut(pin) for pin in self.COL_PINS]
        self.rows = [digitalio.DigitalInOut(pin) for pin in self.ROW_PINS]
        for col in self.cols:
            col.direction = digitalio.Direction.OUTPUT
            col.value = False  
        for row in self.rows:
            row.direction = digitalio.Direction.OUTPUT
            row.value = True  

    def clear(self):
        '''Turn off all LEDs'''
        for col in self.cols: 
            col.value = False 
        for row in self.rows: # Turn off all LEDs
            row.value = True  

    def light_pixel(self, x, y, duration=0.1):
        self.cols[x].value = True
        self.rows[y].value = False

matrix = LEDMatrix5x5()

for i in range(10):
    x = random.randint(0,4)
    y = random.randint(0,4)
    matrix.light_pixel(x,y)
    print("x,y = ", x, " ,  ",y)
    time.sleep(0.7)

matrix.clear()

for i in range(10):
    x = random.randint(0,4)
    y = random.randint(0,4)
    matrix.light_pixel(x,y)
    print("x,y = ", x, " ,  ",y)
    time.sleep(0.7)

matrix.clear()

for i in range(10):
    x = random.randint(0,4)
    y = random.randint(0,4)
    matrix.light_pixel(x,y)
    print("x,y = ", x, " ,  ",y)
    time.sleep(0.7)

matrix.clear()

# Example usage:
# matrix = LEDMatrix5x5()
# matrix.light_pixel(2, 3)  # Light up pixel at (2,3) for 0.1s

```
